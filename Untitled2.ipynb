```python
1.
   A Function is defined using def and can exist independently.

   A Method is A function that is associated with an object.

   Example:

     def greet(name):
         return f"Hello, {name}"

     name = "Alice"
     print(greet(name))            # Function
     print(name.upper())           # Method

2.
   Parameters are the variables listed in a function definition.

   Arguments are the actual values passed to a function.

   Example:

       def add(a, b):  # a, b are parameters
           return a + b

       result = add(3, 5)  # 3, 5 are arguments

3.
   Define a function using def or lambda.

   Call a function by using the function name followed by parentheses.

   Example:

       def square(x):
           return x * x

       print(square(4))

       cube = lambda x: x**3
       print(cube(3))

4.
   Returns a value to the caller and ends the function execution.

   Example:

   def multiply(x, y):
       return x * y

   result = multiply(2, 5)
   print(result)  # 10

5.
   Iterator: Its An object with __next__() method.

   Difference: Iterators maintain state and can be manually iterated using next().

   Example:

   lst = [1, 2, 3]  # Iterable
   it = iter(lst)   # Iterator

   print(next(it))  # 1
   print(next(it))  # 2

6.
   A type of iterable using yield to return values.
   Its being defined like functions but use yield instead of return.

   Example:

   def countdown(n):
       while n > 0:
           yield n
           n -= 1

   for i in countdown(3):
       print(i)  # 3, 2, 1

7.
   advantages of using generatos

   1.Memory-efficient.

   2. Useful for large data or streams.

   3. Lazy evaluation.

8.
   A lambda function in Python is a small, anonymous function defined using the
   lambda keyword.

   Its Used for simple operations, often with map(), filter(), etc.

   Example:

   square = lambda x: x**2
   print(square(4))  # 16

9.
   The map() function in Python applies a given function to each item of an
   iterable.

   Example:

   nums = [1, 2, 3]
   result = list(map(lambda x: x*2, nums))
   print(result)  # [2, 4, 6]

10.
   map(): It Applies a function to each element.

   filter(): It Filters elements based on a condition.

   reduce(): It Applies a function cumulatively to the items of an iterable,
             to reduce the iterable to a single value.

   Example:

   from functools import reduce

   nums = [1, 2, 3, 4]

   print(list(map(lambda x: x*2, nums)))        # [2, 4, 6, 8]
   print(list(filter(lambda x: x%2==0, nums)))  # [2, 4]
   print(reduce(lambda x, y: x + y, nums))      # 10
```
